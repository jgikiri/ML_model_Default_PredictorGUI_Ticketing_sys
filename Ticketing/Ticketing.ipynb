```python
from flask import Flask, render_template, request, redirect, url_for, flash
from flask_mail import Mail, Message

app = Flask(__name__, template_folder= 'location:\\Users\\name\\AppData\\Local\\Programs\\Python\\Python version xxx\\Lib\\site-packages\\flask\\templates')  # Use a single underscore around '__name__'

# Configure email settings
app.config['MAIL_SERVER'] = 'smtp.gmail.com'
app.config['MAIL_PORT'] = 587
app.config['MAIL_USE_TLS'] = True
app.config['MAIL_USERNAME'] = 'XXXXX@gmail.com'
app.config['MAIL_PASSWORD'] = 'XYz412%'  # Use an app-specific password
app.config['MAIL_DEFAULT_SENDER'] = 'XXXXX@gmail.com@gmail.com'

mail = Mail(app)

# Tickets list 
tickets = []




# Home page
@app.route('/')
def home():
    return render_template('index.html', tickets=tickets)

# Create a new ticket
@app.route('/create_ticket', methods=['POST'])
def create_ticket():
    title = request.form.get('title')
    description = request.form.get('description')

    if title and description:
        ticket = {'title': title, 'description': description}
        tickets.append(ticket)
        
        # Send an email notification
        subject = f'New Ticket: {title}'
        message = f'New ticket created:\nTitle: {title}\nDescription: {description}'
        send_email('XXXXX@gmail.com', subject, message)

        flash('Ticket created successfully!', 'success')
    else:
        flash('Title and description are required.', 'error')

    return redirect('/')



# Send an email notification
def send_email(to_email, subject, message):
    msg = Message(subject, recipients=[to_email])
    msg.body = message
    mail.send(msg)

import os

if __name__ == "__main__":
    port = int(os.environ.get('PORT', 5000))
    app.run(host='127.0.0.1', port=port)


```
